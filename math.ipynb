```python
from pandas import *


strs= '4.1 1.5 10.4    5.9     3.4    5.7    1.6    6.1      3.0    3.7 31      4.8     2.0     14.8   5.4   4.2     3.9    4,1     11.1   3.5 4.1    4.1      8.8      5.6    4.3    3.3     7.1    10.3   6.2    7.6 10.8   2.8     9.5     12.9  12.1  0.7     4.0      9.2     4.4   5.7 7.7     6.1      5.7     5.9     4.7   3.9      3.7     3.1     6.1   3.1'
strs=strs.split()
of= Series(strs).value_counts()



df = DataFrame({"Frecuencia": of.values},
               index= of.index)
df["FrecuenciaAcum"] = df["Frecuencia"].cumsum()
df["Relativa"] = df["Frecuencia"].div(df["Frecuencia"].sum())
df["RelativaAcum"] = df["Relativa"].cumsum()
df["Porcentaje"] = df["Relativa"].mul(100)
df["Relativa"] = df["Relativa"].round(2)
df["RelativaAcum"] = df["RelativaAcum"].round(2)
df
# ax=df.plot(y="FrecuenciaAcum",kind="line",color="green")
#ax

```
